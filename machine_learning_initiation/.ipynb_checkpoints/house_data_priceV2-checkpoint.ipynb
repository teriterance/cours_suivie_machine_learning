```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#Lecture du fichier house_data.csv
D = pd.read_csv('house_data.csv')

D.head(5)
```

```python
len(D), max(D['price'])
```

```python
#Commençons par éliminer les outliers
D = D[D['price'] <= 10000]

#Puis les valeurs manquantes
D = D.dropna()

len(D)
```

```python
D['arrondissement'].unique()
```

```python
#Utilisons la fonction train_test_split pour déterminer un échantillon d'entrainement et un échantillon test

u = D['surface'].values.reshape(814,1)

v = D['arrondissement'].values.reshape(814,1)

x = np.concatenate((u,v),axis=1)

y = D['price'].values

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
len(x_train), len(x_test), len(y_train), len(y_test)
```

```python
x_train_regsimple = x_train[:,0].reshape(651,1)
x_test_regsimple = x_test[:,0].reshape(163,1)
```

```python
#Commençons par crée notre premier modèle: une régression simple

model_simple = LinearRegression()
model_simple.fit(x_train_regsimple,y_train)
```

```python
#Calcul des points de quelques points de la droite de régression

u_simple_predict = np.linspace(0,300,200).reshape(200,1)
y_simple_predict = model_simple.predict(u_simple_predict)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(15,10))
plt.scatter(x_train_regsimple,y_train)
plt.plot(u_simple_predict,y_simple_predict, color='red', linestyle='--')

plt.xlabel("Surface en m²", color="Darkblue", fontsize="17")

plt.ylabel("Prix du loyer", color="Darkblue", fontsize="17")

plt.title("Régression simple", color="Darkblue", fontsize="20")
```

```python
model_simple.coef_
```

```python
model_simple.intercept_
```

```python
#Calcul du coefficient de détermination pour tester la qualité de la régression
y_simple_global_predict = model_simple.predict(x[:,0].reshape(814,1))
r2_score(y,y_simple_global_predict)
```

```python
#Prenons maintenant en compte la deuxième variable arrondissement et effectuons une régression multiple

#Affichons notre nuage de points
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,10))
ax = Axes3D(fig)
ax = plt.axes(projection="3d")
ax.scatter(x_train[:,0],x_train[:,1],y_train)
ax.view_init(elev=10,azim=55)
ax.set_xlabel("Surface en m²", fontsize="17")
ax.set_ylabel("Arrondissement", fontsize="17")
ax.set_zlabel("Prix du loyer", fontsize="17")
```

```python
model1 = LinearRegression()

model1.fit(x_train,y_train)
```

```python
#Calculons quelques points de notre plan de régression

u_model1_predict = np.linspace(10,200,200)
v_model1_predict = np.linspace(1,10,200)
```

```python
#Traçons notre plan de régression ainsi que notre nuage de points

X, Y = np.meshgrid(u_model1_predict, v_model1_predict)
Z = model1.intercept_ + model1.coef_[0]*X + model1.coef_[1]*Y
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection="3d")
ax.scatter(D['surface'].values,D['arrondissement'].values,y)
ax.plot_surface(X,Y,Z, rstride=1, cstride=1,antialiased=True)
ax.view_init(elev=5,azim=20)

ax.set_xlabel("Surface en m²", fontsize="17")
ax.set_ylabel("Arrondissement", fontsize="17")
ax.set_zlabel("Prix du loyer", fontsize="17")

ax.set_title("Régression multiple", fontsize=20)
```

```python
model1.coef_
```

```python
model1.intercept_
```

```python
#Calculons le coefficient de détermination pour évaluer la qualité de la régression

y_model1_global_predict = model1.predict(x)
r2_score(y,y_model1_global_predict)
```

```python
#Exemple: évaluation du prix d'un 30m² dans le 10ème arrondissement

model1.predict(np.array([[30,10]]))
```

```python
#On s'apperçoit que le modèle de régression multiple est de meilleur qualité que le modèle de régression simple
#Effecutons maintenant une régression multiple polynomiale, ce qui constitue notre 3ème modèle

poly = PolynomialFeatures(degree=2)

X_train = poly.fit_transform(x_train)

model = LinearRegression()
model.fit(X_train,y_train)
```

```python
#Calcul de quelques points de notre surface de régression

u_model_predict = np.linspace(10,200,200)
v_model_predict = np.linspace(1,10,200)
```

```python
#Traçons notre nuage de points ainsi que notre surface de régression

X, Y = np.meshgrid(u_model_predict, v_model_predict)
C = model.coef_
Z = C[1]*X + C[2]*Y + C[3]*(X**2) + C[4]*(X*Y) + C[5]*(Y**2) + model.intercept_
fig = plt.figure(figsize=(15,15))
ax = plt.axes(projection="3d")
ax.scatter(D['surface'].values,D['arrondissement'].values,y)
ax.plot_surface(X,Y,Z, rstride=1, cstride=1,antialiased=True)
ax.view_init(elev=10,azim=10)

ax.set_xlabel("Surface en m²", fontsize="17")
ax.set_ylabel("Arrondissement", fontsize="17")
ax.set_zlabel("Prix du loyer", fontsize="17")

ax.set_title("Régression multiple polynomiale de degré 2", fontsize=20)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#Calculons le coefficient de détermination pour évaluer la qualité de la régression

X_ = poly.fit_transform(x)
y_model_global_predict = model.predict(X_)
r2_score(y,y_model_global_predict)
```

```python
#Exemple: Prédiction de la valeur du loyer pour un 50 m² dans le 1er arrondissement

L = poly.fit_transform(np.array([[50,1]]))
model.predict(L)
```

```python
#On s'aperçoit que les deux modèles de régression multiples sont supérieurs au modèle de régression simple
#Le modèle qu'il faut choisir est le modèle de régression multiple polynomial!

```
